```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Coast', y='Tempo1', palette={'east_coast': 'red', 'west_coast': 'blue'})
plt.title('Comparison of Tempo by Coast')
plt.show()

# Example: Comparing MFCC Mean Values by Coast
mfcc_columns = [col for col in df.columns if 'MFCC_Mean' in col]
df_mfcc_means = df[['Coast'] + mfcc_columns]
df_mfcc_means_melted = df_mfcc_means.melt(id_vars='Coast', var_name='MFCC', value_name='Mean Value')

plt.figure(figsize=(12, 8))
sns.barplot(data=df_mfcc_means_melted, x='MFCC', y='Mean Value', hue='Coast', palette={'east_coast': 'red', 'west_coast': 'blue'})
plt.xticks(rotation=90)
plt.title('Comparison of MFCC Mean Values by Coast')
plt.show()
```

```python
topic_counts = df.groupby(['Coast', 'Topic Model']).size().unstack('Coast').fillna(0)

topic_counts.plot(kind='bar', figsize=(12, 6))
plt.title('Frequency of Topics by Coast')
plt.xlabel('Topics')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.legend(title='Coast')
plt.show()
```

```python

features = ['Tempo1', 'Duration (s)'] + [col for col in df.columns if 'MFCC_Mean' in col][:10]
correlation_matrix = df[features].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of Selected Features')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'df' is your DataFrame and it includes 'Tempo1' and 'Topic Model' columns
plt.figure(figsize=(12, 8))
sns.boxplot(x='Topic Model', y='Tempo1', data=df)
plt.title('Distribution of Tempo by Topic Model')
plt.xlabel('Topic Model')
plt.ylabel('Tempo')
plt.xticks(rotation=45)  # Rotate the labels to fit them better
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assume 'df' is your DataFrame and it's already been loaded
# Calculate mean tempo for each topic
mean_tempos = df.groupby('Topic Model')['Tempo1'].mean().reset_index()

# Create a repeating color list
colors = ['red', 'blue'] * (len(mean_tempos) // 2 + 1)  # Ensures the list is long enough

plt.figure(figsize=(12, 8))
sns.barplot(x='Topic Model', y='Tempo1', data=mean_tempos, palette=colors)
plt.title('Average Tempo by Topic Model')
plt.xlabel('Topic Model')
plt.ylabel('Average Tempo')
plt.xticks(rotation=45)  # Rotate the labels for better readability
plt.show()
```

```python
# Group by 'Release Year' and 'Topic Model', then calculate the mean tempo
tempo_per_topic_per_year = df.groupby(['Release Year', 'Topic Model'])['Tempo1'].mean().reset_index()
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x='Release Year', y='Tempo1', hue='Topic Model', data=tempo_per_topic_per_year, marker='o')
plt.title('Tempo per Topic Over Time')
plt.xlabel('Release Year')
plt.ylabel('Average Tempo')
plt.legend(title='Topic Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'tempo_per_topic_per_year' has been previously calculated
# and it contains 'Release Year', 'Topic Model', and 'Tempo1' columns

# Define a custom color palette for the topic models
palette = {0: 'red', 1: 'blue'}  # Add more topics and colors as needed

plt.figure(figsize=(14, 8))
sns.lineplot(
    x='Release Year',
    y='Tempo1',
    hue='Topic Model',
    data=tempo_per_topic_per_year,
    marker='o',
    palette=palette  # Use the custom palette
)
plt.title('Tempo per Topic Over Time')
plt.xlabel('Release Year')
plt.ylabel('Average Tempo')
plt.legend(title='Topic Model', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
